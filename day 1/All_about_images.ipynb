```python
from PIL import Image
```

```python
im =Image.open("image.png")
```

```python
new_size=(300,300)
img_resized=im.resize(new_size)

print(img_resized.size)
img_resized
```

```python
import numpy as np
s=np.array(im)
# s
```

```python
#Brighness of image
from PIL import ImageEnhance
enhance = ImageEnhance.Brightness(img_resized)
img_brighten = enhance.enhance(1.6)
img_brighten
```

```python
#Gray scale 
gray_image=img_resized.convert("L")
gray_image
```

```python
#Rotate the image

rotate_img = img_resized.rotate(180)

rotate_img
```

```python
import matplotlib.pyplot as plt
npar=np.array(img_resized)
plt.imshow(npar)
```

```python
#Crop image
# area=(left,top, right , bottom)
area=(100,100,200,200)
crop_image =img_resized.crop(area)
crop_image
```

```python
from PIL import ImageFilter
gray_image=img_resized.convert("L")
edge_image = gray_image.filter(ImageFilter.FIND_EDGES)
edge_image
```

# Edge Detection


```python
from PIL import ImageFilter
gray_image=img_resized.convert("L")
edge_image = gray_image.filter(ImageFilter.Kernel((3,3) ,(-1,-1,-1,-1,8,-1,-1,-1,-1),1,0))
edge_image
```

```python
from PIL import ImageDraw , ImageFont
watermark_image=img_resized.copy()
draw = ImageDraw.Draw(watermark_image)
font = ImageFont.truetype("arial.ttf",40)
draw.text((0,0),"Sachin",(0,255,0),font=font)
watermark_image
```
